```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bn_df = pd.read_csv("Store/BN_training.csv")
ln_df = pd.read_csv("Store/LN_training.csv")
gn_df = pd.read_csv("Store/GN_training.csv")
```

```python
bn_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].plot(bn_df['Train Loss'], label='Batch Norm')
axs[0, 0].set_title("Training Loss")
axs[1, 0].plot(bn_df['Train Acc'], label='Batch Norm')
axs[1, 0].set_title("Training Accuracy")
axs[0, 1].plot(bn_df['Test Loss'], label='Batch Norm')
axs[0, 1].set_title("Test Loss")
axs[1, 1].plot(bn_df['Test Acc'], label='Batch Norm')
axs[1, 1].set_title("Test Accuracy")

axs[0, 0].plot(ln_df['Train Loss'], label='Layer Norm')
axs[1, 0].plot(ln_df['Train Acc'], label='Layer Norm')
axs[0, 1].plot(ln_df['Test Loss'], label='Layer Norm')
axs[1, 1].plot(ln_df['Test Acc'], label='Layer Norm')

axs[0, 0].plot(gn_df['Train Loss'], label='Group Norm')
axs[1, 0].plot(gn_df['Train Acc'], label='Group Norm')
axs[0, 1].plot(gn_df['Test Loss'], label='Group Norm')
axs[1, 1].plot(gn_df['Test Acc'], label='Group Norm')

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
plt.plot()
```
